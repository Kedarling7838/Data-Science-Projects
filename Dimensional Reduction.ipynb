# Applying PCA

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.decomposition import PCA
pca_model=PCA(n_components=0.95)
X_train_pca=pca_model.fit_transform(X_train)
X_test_pca=pca_model.transform(X_test)
```

```python
X_train_pca.shape
```

```python
pca_model.explained_variance_ratio_
```

```python
from sklearn.linear_model import LogisticRegression
clr=LogisticRegression()
clr.fit(X_train_pca,y_train)
```

```python
ypred_train=clr.predict(X_train_pca)
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,X_train_pca,y_train,cv=5).mean())
from sklearn.metrics import accuracy_score
print(accuracy_score(ypred_train,y_train))
```

```python
ypred_test=clr.predict(X_test_pca)
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,X_test_pca,y_test,cv=5).mean())
from sklearn.metrics import accuracy_score
print(accuracy_score(ypred_test,y_test))
```
