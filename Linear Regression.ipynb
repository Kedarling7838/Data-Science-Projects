# 1.Example

```python
import pandas as pd
df=pd.DataFrame({'in':[1,2,3,4,5],'out':[6,7,8,9,10]})
df
```

# Select X(Input) & y(Output)

```python
X=df[['in']] # X should be dataframe
y=df['out'] # y should be series
```

# Modelling

```python
#import
from sklearn.linear_model import LinearRegression
#save in a model
model=LinearRegression()
#fit on data
model.fit(X,y)    #what will calculate ,How will
```

```python
model.coef_
```

```python
model.intercept_
```

# Prediction

```python
model.predict(X)
```

```python
#substite X value in model(equation)
ypred=model.predict(X)
ypred
```

# Evalvation

```python
from sklearn.metrics import mean_squared_error
#mean_squared_error(actual value,predicted vale)
mean_squared_error(y,ypred)
```

```python
mse=mean_squared_error(y,ypred)
mse
```

```python
from math import sqrt
rmse=sqrt(mse)
rmse
```

```python
print("RMSE of this model is:",rmse)
```
