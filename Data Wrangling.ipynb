**FEATURE SCALING**

***Standerdization*** : z=(x-x_mean)/std

```python
import pandas as pd
df=pd.DataFrame({"x":[1,2,3,4,5]})
df
```

***Standerdization using standerd scalar in sklearn***

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df['x_sc_sk']=sc.fit_transform(df[['x']])
df
```

***Standerdization using pandas***

```python
df['x'].mean()
```

```python
df['x'].std(ddof=0)
```

```python
df['x_sc_sk']=(df['x']-3.0)/1.4142135623730951
df['x_sc_sk']
```

**OR**

```python
df['x_sc_sk']=(df['x']-df['x'].mean())/df['x'].std(ddof=0)
df['x_sc_sk']
```

***Normalization***

***MinMax scaling using minmax scaler in sklearn***

```python
from sklearn.preprocessing import MinMaxScaler
min_max= MinMaxScaler()
df['x_mm_sk']=min_max.fit_transform(df[['x']])
df['x_mm_sk']
```

***MinMax scaling using pandas***

```python
df['x_mm_sk']=(df['x']-df['x'].min())/(df['x'].max()-df['x'].min())
df['x_mm_sk']
```

**FEATURE ENCODING**

```python
df1=pd.DataFrame({"shirtSize":["xl","large","medium","small"]})
df1
```

***Ordinal Encoding***

```python
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["xl","large","medium","small"]])
df1["oe_shirtsize"]=oe.fit_transform(df1[['shirtSize']]).astype('int')
df1["oe_shirtsize"]
```

***mapping***

```python
df1["shirtsize_pandas"]=df1["shirtSize"].map({'small':0,'medium':1,'large':2,'xl':3})
df1["shirtsize_pandas"]
```

***Replace***

```python
df1["shirtsize_pandas"]=df1["shirtSize"].replace({'small':0,'medium':1,'large':2,'xl':3})
df1["shirtsize_pandas"]
```
