**mean absolute deviation**

```python
x=np.array([5,9,16,17,18])
deviation=x-x.mean()
deviation
```

```python
abs_deviation=abs(deviation)
abs_deviation
```

```python
np.mean(abs_deviation)
```

**variance/mean square deviation**

```python
x=np.array([5,9,16,17,18])
deviation=x-x.mean()
square_deviation=deviation*deviation
square_deviation.mean()
```

```python
np.var(x)
```

**Standerd deviation**

```python
#option1
var =np.var(x)
std_deviation=np.sqrt(var)
std_deviation
```

```python
#option 2
np.std(x)
```

**Deviation/sample variance**

```python
s=np.array([2398,1844,1539,1311])
sample_variance=np.var(s)*len(s)/len(s)-1
sample_variance
```

****sample standerd deviation****

```python
sample_standerd_deviation=np.sqrt(sample_variance)
sample_standerd_deviation                                
```
