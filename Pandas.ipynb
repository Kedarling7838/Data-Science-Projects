**pandas DataFrame**

**creating dataframe using dictionary**

```python
df1={'col1':[1,2],'col2':[3,4]}
df1=pd.DataFrame(df1)
df1
```

```python
tempreture_df=pd.DataFrame({'city':['mumbai','pune','chennai','kolkata','Banglore'],'tempreture':[23,34,56,32,45]})
tempreture_df
```

```python
tempreture_df.shape
```

```python
tempreture_df.size
```

**Add column**

```python
tempreture_df['Aqi']=[123,321,234,43,543]
tempreture_df
```

```python
humidity_df=pd.DataFrame({'city':['mumbai','pune','hydrabad','Banglore','kolkata'],'humidity':[12,21,32,3,4]})
humidity_df
```

**combining the dataframes**
--using two methods--
--append method--
--concat method--

```python
df1=tempreture_df.append(humidity_df)
df1
```

```python
df1=tempreture_df.append(humidity_df,ignore_index=True)
df1
```

```python
df=pd.concat([tempreture_df,humidity_df],ignore_index=True)
df
```

**merging the dataframes**

```python
dg=pd.merge(tempreture_df,humidity_df,on='city',how='inner')
dg
```

```python
dh=pd.merge(tempreture_df,humidity_df,on='city',how='outer')
dh
```

**Drop columns from the dataframes**

```python
df=tempreture_df.drop(columns=['city','Aqi'])
df
```

```python
df=tempreture_df.drop(columns=['city'])
df
```
