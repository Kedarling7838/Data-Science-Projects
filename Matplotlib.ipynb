***Histogram***

```python
import matplotlib.pyplot as plt
plt.hist(df['Grade'])
plt.show()
```

```python
plt.hist(df['Grade'],bins=10,range=(0,200))
plt.xticks(list(range(0,200,20)))
plt.xlabel("Grade")
plt.ylabel("How many students")
plt.title("frequency distribution grade")
plt.show()
```

***Box plot***:- use to check give data has outlier or not

```python
plt.boxplot(df['Grade'])
plt.show()
```

```python
Q1=df['Grade'].quantile(0.25)
Q3=df['Grade'].quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+(1.5*IQR)
lower_limit=Q1-(1.5*IQR)
df[(df['Grade']<lower_limit)|(df['Grade']>upper_limit)]
```

***Sub-plot***

```python
plt.figure(figsize=(10,10))
#creating histogram in first cell of the figure having 3 rows 2 columns
plt.subplot(3,2,1)
plt.hist(df['Grade'],bins=10,range=(0,200))
#creating barchart in second cell
plt.subplot(3,2,2)
plt.bar([0,1,2,3],turnover,width=0.4,label='turnover')
plt.bar([0.4,1.4,2.4,3.4],profit,width=0.4,label='profit')
plt.legend()
#creating line chart in third cell
plt.subplot(3,2,3)
plt.plot(df['Grade'],marker="o",color='orange')
#creating scatter plot in 4th cell
plt.subplot(3,2,4)
plt.scatter(df['studt_hours'],df['Grade'],color="red",marker="o")
plt.xlabel('studt_hours')
plt.ylabel('grade')
#creating piechart plot in 5th cell
plt.subplot(3,2,5)
plt.pie(x=[20,25,33],labels=['s','m','l'],autopct='%0.1f%%')
#creating Barchart plot in 6th cell
plt.subplot(3,2,6)
plt.bar(["GOOGL","AMMZN","MSFT","FB"],turnover,label='turnover')
plt.bar(["GOOGL","AMMZN","MSFT","FB"],profit,label='profit')
plt.legend()
plt.show()
```
