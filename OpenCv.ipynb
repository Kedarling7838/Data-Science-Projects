# Face Detection Using HAAR Cascade Classifier

```python
face_classifier=cv2.CascadeClassifier("C:\\Users\\Asus\\OneDrive\\Desktop\\power shheets\\Haarcascades\\haarcascade_frontalface_default.xml")
img=cv2.imread("C:\\Users\\Asus\\OneDrive\\Desktop\\power shheets\\image_examples\\image.jpg",1)
img=cv2.resize(img,(679,750))
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_classifier.detectMultiScale(img,1.5,5)
if faces is ():
    print("No face found")
else:
    for (x,y,l,h) in faces:
        cv2.rectangle(img, (x,y), (x+l,y+h),(255,0,255), 1)

cv2.imshow('Face Detection',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
